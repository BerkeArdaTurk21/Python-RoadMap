## Iterator Cheatsheet

| Concept | What it means |
|---|---|
| Iterable | Has `__iter__()` — can be looped over (list, str, range...) |
| Iterator | Has `__iter__()` AND `__next__()` — tracks position |
| `iter(obj)` | Calls `obj.__iter__()`, returns an iterator |
| `next(it)` | Calls `it.__next__()`, returns next value |
| `StopIteration` | Raised by `__next__` when no items remain |
| `next(it, default)` | Returns `default` instead of raising `StopIteration` |
| `__iter__` in iterator | Always returns `self` |

```python
# 1. Iterable vs Iterator

numbers = [1, 2, 3]       # list is ITERABLE (not an iterator)
it = iter(numbers)         # iter() creates an ITERATOR

print(type(numbers))       # <class 'list'>
print(type(it))            # <class 'list_iterator'>

print(next(it))            # 1
print(next(it))            # 2
print(next(it))            # 3
print(next(it, 'done'))    # done  (no StopIteration)
```

```python
# 2. Custom iterator — Countdown

class Countdown:
    def __init__(self, start):
        self.current = start

    def __iter__(self):
        return self          # iterator returns itself

    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        value = self.current
        self.current -= 1
        return value

# Use next() manually:
cd = Countdown(3)
print(next(cd))   # 3
print(next(cd))   # 2
print(next(cd))   # 1

# Use in a for loop:
for n in Countdown(5):
    print(n, end=' ')   # 5 4 3 2 1
```

```python
# 3. How for loops work internally

fruits = ['apple', 'banana', 'cherry']

# This for loop:
for fruit in fruits:
    print(fruit)

print('---')

# Is equivalent to:
_it = iter(fruits)
while True:
    try:
        fruit = next(_it)
        print(fruit)
    except StopIteration:
        break
```

```python
# 4. Custom iterator — Fibonacci

class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

print(list(Fibonacci(8)))   # [0, 1, 1, 2, 3, 5, 8, 13]

# Once exhausted, cannot be reused:
fib = Fibonacci(5)
print(list(fib))    # [0, 1, 1, 2, 3]
print(list(fib))    # []  <- exhausted
```

```python
# 5. Infinite iterator + islice
import itertools

class InfiniteCounter:
    def __init__(self, start=0, step=1):
        self.current = start
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        value = self.current
        self.current += self.step
        return value

odds = InfiniteCounter(start=1, step=2)
print(list(itertools.islice(odds, 6)))    # [1, 3, 5, 7, 9, 11]

# Built-in zip, enumerate, map, filter also return iterators:
pairs = zip([1, 2, 3], ['a', 'b', 'c'])
print(next(pairs))   # (1, 'a')
print(list(pairs))   # [(2, 'b'), (3, 'c')]
```

## Summary

| Concept | Key Point |
|---|---|
| Iterable | Has `__iter__` — produces an iterator when passed to `iter()` |
| Iterator | Has `__iter__` + `__next__` — stateful, one-way, exhaustible |
| `__iter__` | Returns `self` in iterator classes |
| `__next__` | Returns next value or raises `StopIteration` |
| For loop | Calls `iter()` then `next()` in a loop until `StopIteration` |
| Infinite iterator | Never raises `StopIteration` — use `islice` to limit |
