## Generator Cheatsheet

| Concept | What it means |
|---|---|
| Generator function | A `def` containing `yield`; calling it returns a generator |
| Lazy | Body runs only as you iterate; one value at a time |
| `yield x` | Pause, hand `x` to the caller, resume on next `next()` |
| `(expr for x in it)` | Generator expression — like a comprehension but lazy |
| `return` | Ends the generator (raises `StopIteration`) |
| `gen.send(v)` | Resume, with `yield` evaluating to `v` (prime with `next` first) |
| `yield from it` | Delegate — yield every item of `it` |

**Iterator (Day 02) vs Generator (Day 03):** both are iterators, but a generator needs no class — just `yield`.

## 1. yield — a lazy function

Calling a generator function returns a generator object; the body runs only as you iterate.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

gen = count_up_to(3)
print(type(gen))
print(next(gen))
print(next(gen))
print(list(count_up_to(5)))   # generators work in list(), sum(), for...
```

## 2. Lazy = memory efficient

A generator stores one value at a time, so it stays tiny even for millions of items — and can be infinite.

```python
import sys

as_list = [x * x for x in range(1_000_000)]
as_gen  = (x * x for x in range(1_000_000))
print('list:', sys.getsizeof(as_list), 'bytes')
print('gen :', sys.getsizeof(as_gen), 'bytes')

def naturals():
    n = 1
    while True:
        yield n
        n += 1

nats = naturals()
print([next(nats) for _ in range(5)])
```

## 3. Generator expressions & pipelines

`(expr for x in it)` is lazy. Chain stages together — nothing is computed until consumed.

```python
print(sum(x * x for x in range(5)))      # parens optional as sole arg

nums    = (n for n in range(20))
evens   = (n for n in nums if n % 2 == 0)
doubled = (n * 2 for n in evens)
print(list(doubled))
```

## 4. send() and yield from

`send()` feeds a value back into the generator; `yield from` delegates to another iterable.

```python
def accumulator():
    total = 0
    while True:
        x = yield total
        if x is not None:
            total += x

acc = accumulator()
next(acc)                 # prime
print(acc.send(10))       # 10
print(acc.send(5))        # 15

def chain(*iterables):
    for it in iterables:
        yield from it

print(list(chain([1, 2], (3, 4), 'ab')))
```
