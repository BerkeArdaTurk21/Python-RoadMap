## Type Hints Cheatsheet

| Hint | Meaning |
|---|---|
| `x: int` | x is expected to be an int |
| `-> str` | function returns a str |
| `list[int]`, `dict[str, int]` | generic collection hints (3.9+) |
| `Optional[X]` / `X \| None` | X or None |
| `Union[X, Y]` / `X \| Y` | X or Y |
| `Callable[[int, int], int]` | a function taking two ints, returning int |
| `TypeVar("T")` + `Generic[T]` | write code that works with any type, type-safely |

> Type hints are **not enforced at runtime** — use `mypy` to check them statically.

```python
# 1. Basic hints

def add(a: int, b: int) -> int:
    return a + b

print(add(2, 3))          # 5
print(add("2", "3"))      # '23' — Python does NOT enforce hints at runtime!

def greet(name: str, excited: bool = False) -> str:
    suffix = "!" if excited else "."
    return f"Hello, {name}{suffix}"

print(greet("Bob", excited=True))
```

```python
# 2. Generic collections (Python 3.9+)

def total(numbers: list[int]) -> int:
    return sum(numbers)

def word_lengths(words: list[str]) -> dict[str, int]:
    return {w: len(w) for w in words}

print(total([1, 2, 3]))
print(word_lengths(["hi", "python"]))

def group_scores(data: dict[str, list[int]]) -> dict[str, float]:
    return {name: sum(scores) / len(scores) for name, scores in data.items()}

print(group_scores({"Alice": [90, 80, 100]}))
```

```python
# 3. Optional and Union
from typing import Optional, Union

def find_user(user_id: int) -> Optional[str]:
    users = {1: "Alice", 2: "Bob"}
    return users.get(user_id)

print(find_user(1))    # Alice
print(find_user(99))   # None

def parse_id(value: Union[int, str]) -> int:
    return int(value)

print(parse_id(42))
print(parse_id("42"))
```

```python
# 4. Callable and TypeVar/Generic
from typing import Callable, TypeVar, Generic

def apply_operation(x: int, y: int, op: Callable[[int, int], int]) -> int:
    return op(x, y)

def multiply(a: int, b: int) -> int:
    return a * b

print(apply_operation(3, 4, multiply))   # 12

T = TypeVar("T")

class Box(Generic[T]):
    def __init__(self, item: T) -> None:
        self.item = item

    def get(self) -> T:
        return self.item

int_box: Box[int] = Box(42)
str_box: Box[str] = Box("hello")
print(int_box.get(), str_box.get())
```

```python
# 5. dataclasses with type hints
from dataclasses import dataclass

@dataclass
class Point:
    x: float
    y: float
    label: str = "origin"

p = Point(1.5, 2.5)
print(p)
print(p.x, p.y)
```

## Summary

| Concept | Key Point |
|---|---|
| Type hints | Documentation + IDE support — NOT enforced at runtime |
| `list[int]`, `dict[str, int]` | Built-in generics since Python 3.9 |
| `Optional[X]` | Shorthand for `Union[X, None]` (or `X \| None` in 3.10+) |
| `Callable[[Args], Ret]` | Hints a function's signature |
| `TypeVar` + `Generic` | Write type-safe code that works with any type |
| `mypy` | Static type checker — run separately, catches mismatches before runtime |
