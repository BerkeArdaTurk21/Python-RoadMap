## 1. Creating Sets

A set is an **unordered** collection of **unique**, **hashable** elements.

> ⚠️ `{}` creates an empty **dict**, not a set. Use `set()` for an empty set!

````python
print(type({}))      # <class 'dict'>  ← trap!
print(type(set()))   # <class 'set'>

numbers = {1, 2, 3, 4, 5}
dupes   = {1, 2, 2, 3, 3, 3}   # duplicates removed automatically
print(numbers)
print(dupes)   # {1, 2, 3}

# Deduplicate a list
names = ['alice', 'bob', 'alice', 'carol']
print(set(names))   # {'alice', 'bob', 'carol'}
````

## 2. Basic Properties

````python
s = {3, 1, 4, 1, 5, 9}

print(len(s))     # 5 (1 duplicate removed)
print(3 in s)     # True  — O(1) lookup
print(7 in s)     # False

# No indexing!
try:
    print(s[0])
except TypeError as e:
    print(e)
````

## 3. Methods — Adding & Removing

````python
fruits = {'apple', 'banana', 'cherry'}

fruits.add('date')       # add one
fruits.add('apple')      # already exists — no change, no error
print(fruits)

fruits.remove('banana')  # KeyError if missing
fruits.discard('mango')  # safe — no error if missing
print(fruits)
````

## 4. Set Operations

```
a = {1,2,3,4,5}    b = {4,5,6,7,8}
       ┌──────────────────────┐
 a only│ 1 2 3 │ both │ 6 7 8│ b only
       │       │  4 5 │      │
       └──────────────────────┘
```

````python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print('union:       ', a | b)   # {1,2,3,4,5,6,7,8}
print('intersection:', a & b)   # {4,5}
print('difference:  ', a - b)   # {1,2,3}
print('sym diff:    ', a ^ b)   # {1,2,3,6,7,8}
````

## 5. Set Comparisons

````python
x = {1, 2, 3}
y = {1, 2, 3, 4, 5}
z = {10, 20}

print(x <= y)          # True  — x is subset of y
print(y >= x)          # True  — y is superset of x
print(x < y)           # True  — strict subset (x != y)
print(x.isdisjoint(z)) # True  — no common elements
print({1,2,3} == {3,2,1})  # True — order doesn't matter
````

## 6. frozenset — Immutable Set

````python
fs = frozenset([1, 2, 3, 4])
print(fs)
print(1 in fs)   # True

# Can be used as a dict key (regular set cannot)
graph = {
    frozenset({1, 2}): 'edge A',
    frozenset({2, 3}): 'edge B',
}
print(graph[frozenset({1, 2})])   # edge A
````

## 7. Practical Use Cases

````python
# Deduplicate a list
data = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
print(list(set(data)))

# Fast membership test
valid = {'alice', 'bob', 'carol'}
print('alice' in valid)   # True — O(1)

# Common elements between two lists
a = [1, 2, 3, 4, 5]
b = [3, 4, 5, 6, 7]
print(set(a) & set(b))    # {3, 4, 5}

# Check for duplicates
def has_duplicates(lst):
    return len(lst) != len(set(lst))

print(has_duplicates([1, 2, 2]))   # True
print(has_duplicates([1, 2, 3]))   # False
````

## Summary

| Concept | Key Point |
|---------|----------|
| Unique | Duplicates silently ignored |
| Unordered | No indexing, no slicing |
| Empty set | `set()` — NOT `{}` |
| `add` / `discard` | `discard` is safe (no error if missing) |
| `\|` union | In either set |
| `&` intersection | In both sets |
| `-` difference | In A but not B |
| `^` sym. difference | In one but not both |
| `frozenset` | Immutable, can be a dict key |
