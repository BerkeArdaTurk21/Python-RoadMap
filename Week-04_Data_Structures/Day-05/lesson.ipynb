## 1. List Comprehensions

**Syntax:** `[expression for item in iterable]`

```python
squares = [x ** 2 for x in range(6)]
print(squares)   # [0, 1, 4, 9, 16, 25]

chars   = [c.upper() for c in 'hello']
print(chars)     # ['H', 'E', 'L', 'L', 'O']

lengths = [len(w) for w in ['apple', 'banana', 'cherry']]
print(lengths)   # [5, 6, 6]
```

## 2. With Condition — Filtering

**Syntax:** `[expression for item in iterable IF condition]`  
The `if` at the **end** filters — shorter list.

```python
evens       = [x for x in range(20) if x % 2 == 0]
even_squares = [x**2 for x in range(10) if x % 2 == 0]

print(evens)
print(even_squares)
```

## 3. if/else — Value Transformation

**Syntax:** `[A if condition else B for item in iterable]`  
`if/else` BEFORE `for` → transforms values, same-length list.

```python
labels  = ['even' if x % 2 == 0 else 'odd' for x in range(6)]
clamped = [x if x >= 0 else 0 for x in [3, -1, 4, -1, 5]]

print(labels)    # ['even', 'odd', 'even', 'odd', 'even', 'odd']
print(clamped)   # [3, 0, 4, 0, 5]
```

## 4. Nested List Comprehensions

```python
matrix = [[1,2,3],[4,5,6],[7,8,9]]

# Flatten — outer loop first, mirrors nested for loops
flat = [cell for row in matrix for cell in row]
print(flat)   # [1,2,3,4,5,6,7,8,9]

# Cartesian product
coords = [(x, y) for x in range(3) for y in range(3)]
print(coords)
```

## 5. Dict Comprehensions

**Syntax:** `{key_expr: value_expr for item in iterable}`

```python
squares_dict = {x: x**2 for x in range(6)}
print(squares_dict)

# Filter
prices = {'apple': 1.2, 'banana': 0.5, 'cherry': 3.0}
cheap  = {k: v for k, v in prices.items() if v < 2}
print(cheap)

# Invert
inv = {v: k for k, v in {'a':1,'b':2,'c':3}.items()}
print(inv)
```

## 6. Set Comprehensions

**Syntax:** `{expression for item in iterable}` — duplicates removed automatically.

```python
unique_lengths = {len(w) for w in ['hi', 'hello', 'hey', 'world', 'sun']}
print(unique_lengths)   # {2, 3, 5}

first_letters = {w[0] for w in ['apple', 'avocado', 'banana', 'cherry']}
print(first_letters)    # {'a', 'b', 'c'}
```

## 7. Generator Expressions

**Syntax:** `(expression for item in iterable)`  
Like a list comprehension but **lazy** — computes one value at a time. Much less memory.

```python
# List comp  → entire list in memory
lst = [x**2 for x in range(1_000_000)]

# Generator  → ~200 bytes regardless of size
gen = (x**2 for x in range(1_000_000))

print(type(lst), type(gen))

# Pass directly to sum/max/any/all — no [] needed
print(sum(x**2 for x in range(10)))         # 285
print(any(x > 50 for x in [10, 30, 55]))    # True
print(all(x > 0 for x in [1, 2, 3]))        # True
```

## Summary

| Type | Syntax | Result |
|------|--------|--------|
| List | `[expr for x in it]` | list |
| List + filter | `[expr for x in it if cond]` | shorter list |
| List + branch | `[a if c else b for x in it]` | same-length list |
| Dict | `{k: v for x in it}` | dict |
| Set | `{expr for x in it}` | set (unique) |
| Generator | `(expr for x in it)` | lazy generator |
| Nested | `[e for a in it1 for b in it2]` | flat list |
