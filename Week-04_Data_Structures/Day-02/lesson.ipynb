## 1. Creating Tuples

A tuple is an **ordered, immutable** sequence. Once created it cannot change.

> ⚠️ A single-element tuple needs a trailing comma: `(42,)` — without it `(42)` is just an `int`!

```python
single    = (42,)       # tuple
not_tuple = (42)        # int!

print(type(single))     # <class 'tuple'>
print(type(not_tuple))  # <class 'int'>

# Parentheses are optional — the COMMA creates the tuple
t = 1, 2, 3
print(t, type(t))       # (1, 2, 3) <class 'tuple'>
```

## 2. Indexing & Slicing

Same rules as lists — zero-indexed, negative indexes count from the end.

```python
coords = (10, 20, 30, 40, 50)

print(coords[0])     # 10
print(coords[-1])    # 50
print(coords[1:4])   # (20, 30, 40)
print(coords[::-1])  # (50, 40, 30, 20, 10)
```

## 3. Immutability

```python
point = (3, 7)

# This raises TypeError:
try:
    point[0] = 99
except TypeError as e:
    print(e)   # 'tuple' object does not support item assignment

# BUT a mutable object INSIDE a tuple can change:
t = ([1, 2], [3, 4])
t[0].append(99)
print(t)   # ([1, 2, 99], [3, 4])  — the list changed, the tuple structure did not
```

## 4. Packing & Unpacking

```python
# Packing
person = 'Berke', 21, 'Warsaw'
print(person)   # ('Berke', 21, 'Warsaw')

# Unpacking
name, age, city = person
print(name, age, city)

# Extended unpacking with *
first, *rest = (1, 2, 3, 4, 5)
print(first, rest)   # 1 [2, 3, 4, 5]

a, *middle, b = (1, 2, 3, 4, 5)
print(a, middle, b)  # 1 [2, 3, 4] 5
```

## 5. Swap Without Temp Variable

```python
x, y = 10, 20
x, y = y, x     # tuple packing + unpacking in one step
print(x, y)     # 20 10
```

## 6. Functions Returning Multiple Values

```python
def min_max(numbers):
    return min(numbers), max(numbers)   # actually returns a tuple

result = min_max([4, 1, 7, 2, 9])
print(result, type(result))    # (1, 9) <class 'tuple'>

low, high = min_max([3, 5, 1, 8])  # unpack directly
print(low, high)
```

## 7. Named Tuples

Give each position a **name** — like a lightweight, immutable class.

```python
from collections import namedtuple

Point  = namedtuple('Point',  ['x', 'y'])
Person = namedtuple('Person', ['name', 'age', 'city'])

p   = Point(3, 7)
bob = Person('Bob', 30, 'London')

print(p)          # Point(x=3, y=7)
print(p.x, p.y)   # 3  7
print(p[0])       # 3  — index still works

print(bob.name)   # Bob

# _replace returns a NEW tuple with changes
p2 = p._replace(x=99)
print(p2)         # Point(x=99, y=7)
```

## 8. Tuples as Dict Keys

```python
# Tuples are hashable → can be used as dict keys
# Lists are NOT hashable → TypeError if used as dict keys

locations = {
    (40.7128, -74.0060): 'New York',
    (51.5074, -0.1278):  'London',
    (52.2297, 21.0122):  'Warsaw',
}
print(locations[(40.7128, -74.0060)])  # New York
```

## Summary

| Concept | Key Point |
|---------|----------|
| Immutable | Cannot change after creation |
| Single-item | `(42,)` — comma is mandatory |
| Packing | `t = 1, 2, 3` |
| Unpacking | `a, b, c = t` |
| Extended `*` | `first, *rest = t` |
| Swap | `x, y = y, x` |
| namedtuple | Named fields, still immutable |
| Dict key | Tuples can; lists cannot |
