```python
from collections import Counter, defaultdict, namedtuple, deque, OrderedDict
```

## 1. Counter

Counts hashable objects. Missing keys return **0** (no KeyError).

```python
words  = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']
c      = Counter(words)
print(c)

print(c['apple'])    # 3
print(c['mango'])    # 0  — no KeyError!

print(c.most_common(2))   # [('apple', 3), ('banana', 2)]

# From string
print(Counter('mississippi'))
```

```python
# Counter arithmetic
a = Counter({'x': 3, 'y': 2})
b = Counter({'x': 1, 'y': 4, 'w': 2})

print('a + b:', a + b)   # add counts
print('a - b:', a - b)   # subtract, drop zeros/negatives
print('a & b:', a & b)   # min of each
print('a | b:', a | b)   # max of each
```

## 2. defaultdict

Like a dict but returns a **default value** for missing keys — no KeyError.

```python
# defaultdict(list) — auto-creates [] for new keys
groups = defaultdict(list)
data   = [('fruit','apple'), ('veggie','carrot'), ('fruit','banana')]

for category, item in data:
    groups[category].append(item)   # no if-check needed!

print(dict(groups))
```

```python
# defaultdict(int) — auto-creates 0 → perfect for counting
word_count = defaultdict(int)
for word in 'the quick brown fox the fox'.split():
    word_count[word] += 1

print(dict(word_count))
print(word_count['missing'])   # 0
```

## 3. namedtuple

```python
Student = namedtuple('Student', ['name', 'age', 'gpa'])

students = [
    Student('Alice', 20, 3.8),
    Student('Bob',   22, 3.2),
    Student('Carol', 21, 3.9),
]

for s in students:
    print(f'{s.name}: GPA {s.gpa}')

top = max(students, key=lambda s: s.gpa)
print('Top student:', top.name)

# _replace returns a new instance
updated = students[0]._replace(gpa=4.0)
print(updated)
```

## 4. deque (double-ended queue)

`list.pop(0)` is **O(n)** — shifts every element.  
`deque.popleft()` is **O(1)** — that's the whole point.

```python
dq = deque([1, 2, 3])
dq.appendleft(0)   # O(1)
dq.append(4)       # O(1)
print(dq)

dq.popleft()       # O(1)
print(dq)

# rotate
dq2 = deque([1, 2, 3, 4, 5])
dq2.rotate(2)
print(dq2)   # deque([4, 5, 1, 2, 3])
```

```python
# maxlen — fixed-size sliding window
recent = deque(maxlen=3)
for x in range(6):
    recent.append(x)
    print(list(recent))
```

## 5. OrderedDict

```python
od = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
od.move_to_end('a')           # move to end
print(od)

od.move_to_end('a', last=False)  # move to front
print(od)

# Order matters for equality
od1 = OrderedDict([('a',1),('b',2)])
od2 = OrderedDict([('b',2),('a',1)])
print(od1 == od2)   # False
```

## Summary

| Class | Use case |
|-------|----------|
| `Counter` | Count occurrences; top-N; arithmetic on counts |
| `defaultdict` | Grouping, counting — no KeyError on new keys |
| `namedtuple` | Lightweight immutable record with named fields |
| `deque` | Fast O(1) append/pop at both ends; sliding windows |
| `OrderedDict` | `move_to_end()`; order-sensitive equality |
