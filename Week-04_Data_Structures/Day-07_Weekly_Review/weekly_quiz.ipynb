## Part 1 — Multiple Choice

### Q1. List Slicing
```python
lst = [10, 20, 30, 40, 50]
print(lst[1:4:2])
```
- A) `[20, 30, 40]`  
- B) `[20, 40]`  
- C) `[10, 30, 50]`  
- D) `[30, 40]`

````python
# Answer: B — start=1(20), step=2 → index 1(20), 3(40). Stop before index 4.
lst = [10, 20, 30, 40, 50]
print(lst[1:4:2])   # [20, 40]
````

### Q2. List Reference Trap
```python
a = [1, 2, 3]
b = a
b.append(4)
# What are a and b?
```
- A) `a=[1,2,3,4]  b=[1,2,3,4]`  
- B) `a=[1,2,3]    b=[1,2,3,4]`

````python
# Answer: A — b = a is a REFERENCE, not a copy!
a = [1, 2, 3]
b = a
b.append(4)
print('a:', a)   # [1, 2, 3, 4]
print('b:', b)   # [1, 2, 3, 4]

# Fix: use b = a.copy() or b = a[:]
````

### Q3. Single-Element Tuple
```python
x = (42)
print(type(x))
```
- A) tuple  B) **int** ← correct  C) list  D) set

````python
print(type((42)))    # <class 'int'>
print(type((42,)))   # <class 'tuple'>  ← comma makes it a tuple
````

### Q4. Dict Comprehension Inversion
```python
d = {'a': 1, 'b': 2, 'c': 3}
result = {v: k for k, v in d.items()}
print(result[2])
```
- A) KeyError  B) 'a'  C) **'b'** ← correct  D) 2

````python
d = {'a': 1, 'b': 2, 'c': 3}
result = {v: k for k, v in d.items()}
print(result)     # {1:'a', 2:'b', 3:'c'}
print(result[2])  # 'b'
````

### Q5. Symmetric Difference
```python
a = {1,2,3,4};  b = {3,4,5,6}
print(a ^ b)
```
- A) {3,4}  B) **{1,2,5,6}** ← correct  C) {1,2,3,4,5,6}  D) {1,2}

````python
a = {1,2,3,4}
b = {3,4,5,6}
print(a ^ b)   # {1, 2, 5, 6} — in one but not both
````

### Q6 — Q10 (verify in code)

````python
# Q6: empty set
print(type({}))      # dict!
print(type(set()))   # set

# Q7: list comp with filter
print([x**2 for x in range(6) if x % 2 == 0])   # [0, 4, 16]

# Q8: generator vs list comp
import sys
lst = [x**2 for x in range(1000)]
gen = (x**2 for x in range(1000))
print('list:', sys.getsizeof(lst), 'bytes')
print('gen: ', sys.getsizeof(gen), 'bytes')

# Q9: Counter missing key
from collections import Counter
c = Counter('banana')
print(c['z'])   # 0

# Q10: deque popleft O(1)
from collections import deque
dq = deque([1,2,3,4,5])
dq.popleft()   # O(1)
print(list(dq))
````

## Part 2 — Code Challenges

````python
# C1: Extended unpacking
first, *middle, last = (10, 20, 30, 40, 50)
print(first + last)   # 60
print(middle)         # [20, 30, 40]  — a LIST!
````

````python
# C2: Dict merge — right side wins
d1 = {'a': 1, 'b': 2}
d2 = {'b': 99, 'c': 3}
merged = {**d1, **d2}
print(sorted(merged.items()))   # [('a',1), ('b',99), ('c',3)]
````

````python
# C3: Set comprehension — unique lengths
words = ['hi', 'hello', 'hey', 'world']
result = {len(w) for w in words}
print(sorted(result))   # [2, 3, 5]
````

````python
# C4: defaultdict counting
from collections import defaultdict
d = defaultdict(int)
for ch in 'abracadabra':
    d[ch] += 1
print(d['a'])   # 5
````

````python
# C5: Flatten + filter + generator sum
matrix = [[1,2],[3,4],[5,6]]
flat   = [n for row in matrix for n in row]
print(flat)
print(sum(x for x in flat if x % 2 == 0))   # 2+4+6 = 12
````
