## 1. Creating Dictionaries

A dictionary is an **ordered** (Python 3.7+), **mutable** collection of `key → value` pairs.  
Keys must be **unique** and **hashable** (strings, ints, tuples — not lists).

```python
person = {'name': 'Berke', 'age': 21, 'city': 'Warsaw'}
print(person)
print(type(person))

# dict() constructor
d = dict(name='Bob', age=30)
print(d)
```

## 2. Accessing Values

| Method | Missing key behaviour |
|--------|-----------------------|
| `d['key']` | Raises `KeyError` |
| `d.get('key')` | Returns `None` |
| `d.get('key', default)` | Returns `default` |

```python
person = {'name': 'Berke', 'age': 21}

print(person['name'])                    # Berke
print(person.get('email'))               # None
print(person.get('email', 'N/A'))        # N/A
print('name' in person)                  # True
print('email' in person)                 # False
```

## 3. Modifying Dictionaries

```python
d = {'a': 1, 'b': 2}

d['c'] = 3              # add
d['a'] = 99             # update
d.update({'b': 50, 'd': 4})  # update multiple
print(d)

val = d.pop('d')        # remove + return
print('popped:', val)
del d['c']              # delete
print(d)
```

## 4. Key Methods

```python
inv = {'apple': 5, 'banana': 3, 'cherry': 8}

print(list(inv.keys()))    # ['apple', 'banana', 'cherry']
print(list(inv.values()))  # [5, 3, 8]
print(list(inv.items()))   # [('apple',5), ('banana',3), ('cherry',8)]

# setdefault — adds key only if missing
inv.setdefault('date', 0)
print(inv['date'])   # 0
inv.setdefault('apple', 999)
print(inv['apple'])  # 5 — unchanged
```

## 5. Iterating

```python
prices = {'apple': 1.20, 'banana': 0.50, 'cherry': 2.00}

for key, value in prices.items():
    print(f'{key}: ${value:.2f}')
```

## 6. Nested Dictionaries

```python
students = {
    'alice': {'age': 20, 'grade': 'A', 'scores': [95, 88, 92]},
    'bob':   {'age': 22, 'grade': 'B', 'scores': [78, 82, 80]},
}

print(students['alice']['grade'])       # A
print(students['bob']['scores'][1])     # 82

# Safe nested access
email = students.get('alice', {}).get('email', 'no email')
print(email)

# Iterate + compute average
for name, info in students.items():
    avg = sum(info['scores']) / len(info['scores'])
    print(f"{name}: avg={avg:.1f}")
```

## 7. Merging Dictionaries

```python
defaults  = {'color': 'blue', 'size': 'M', 'quantity': 1}
overrides = {'color': 'red', 'quantity': 3}

# ** unpacking — right side wins on duplicate keys
merged = {**defaults, **overrides}
print(merged)

# Python 3.9+ pipe operator
merged2 = defaults | overrides
print(merged2)
```

## 8. Word Count Pattern

```python
words = ['apple', 'banana', 'apple', 'cherry', 'banana', 'apple']

counts = {}
for word in words:
    counts[word] = counts.get(word, 0) + 1

print(counts)   # {'apple': 3, 'banana': 2, 'cherry': 1}
```

## 9. Dict Comprehension (preview)

```python
squares = {x: x**2 for x in range(6)}
print(squares)

# Invert a dict
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print(inverted)
```

## Summary

| Operation | Example |
|-----------|--------|
| Safe access | `d.get('key', default)` |
| Add / update | `d['key'] = value` |
| Update many | `d.update({...})` |
| Delete | `del d['key']` / `d.pop('key')` |
| Iterate pairs | `for k, v in d.items()` |
| Check key | `'key' in d` |
| Merge | `{**d1, **d2}` or `d1 \| d2` |
| Word count | `d[w] = d.get(w, 0) + 1` |
