## os Module Cheatsheet

| Function | Purpose |
|---|---|
| `os.getcwd()` | Current working directory |
| `os.listdir(path)` | List names in directory |
| `os.scandir(path)` | List as DirEntry objects (faster) |
| `os.path.join(a, b)` | Build paths safely |
| `os.path.exists(p)` | Check if path exists |
| `os.makedirs(p, exist_ok=True)` | Create directory tree |
| `os.remove(p)` | Delete a file |
| `os.rename(src, dst)` | Rename/move a file |
| `os.walk(top)` | Recursive directory traversal |
| `os.getenv(key)` | Read environment variable safely |
| `os.environ[key]` | Read env var (raises KeyError if missing) |

```python
import os

# Current directory
cwd = os.getcwd()
print(f'cwd:         {cwd}')
print(f'basename:    {os.path.basename(cwd)}')
print(f'parent:      {os.path.dirname(cwd)}')
```

```python
# Path operations
p = os.path.join('projects', 'app', 'main.py')
print(f'join:      {p}')
print(f'exists:    {os.path.exists(".")}')
print(f'isdir:     {os.path.isdir(".")}')

full = '/home/user/data/report.csv'
print(f'dirname:   {os.path.dirname(full)}')
print(f'basename:  {os.path.basename(full)}')
print(f'splitext:  {os.path.splitext(full)}')
```

```python
# Create and remove a directory tree
test = os.path.join(os.getcwd(), '_nb_test', 'sub')
os.makedirs(test, exist_ok=True)
print(f'Created: {test}')
print(f'Exists:  {os.path.isdir(test)}')
os.removedirs(test)
print(f'Removed. Exists: {os.path.exists(test)}')
```

```python
# Environment variables
home = os.getenv('HOME') or os.getenv('USERPROFILE', 'unknown')
print(f'Home: {home}')

venv = os.getenv('VIRTUAL_ENV')
print(f'venv active: {bool(venv)}')

os.environ['DEMO_VAR'] = 'hello'
print(f'DEMO_VAR: {os.getenv("DEMO_VAR")}')
```

```python
# os.walk — recursive traversal
print('Directory tree (depth 1):')
for dirpath, dirnames, filenames in os.walk('.'):
    dirnames[:] = [d for d in dirnames if not d.startswith('.')]
    depth = dirpath.count(os.sep)
    if depth > 1:
        break
    indent = '  ' * depth
    print(f'{indent}{os.path.basename(dirpath) or "."}/')
    for f in filenames[:3]:
        print(f'{indent}  {f}')
```

## Summary

| Concept | Key Point |
|---|---|
| `os.getcwd()` | Absolute path of current dir |
| `os.path.join()` | Always use this to build paths |
| `os.makedirs(exist_ok=True)` | Safe directory creation |
| `os.getenv(key)` | Safer than `os.environ[key]` |
| `os.walk()` | Yields `(dirpath, dirnames, filenames)` |
| `os.name` | `'nt'` = Windows, `'posix'` = Linux/Mac |
