**Topics:** `date` · `time` · `datetime` · `timedelta` · `strftime` · `strptime`

---

## Cheatsheet

| Tool | Purpose |
|---|---|
| `date.today()` | Current calendar date |
| `datetime.now()` | Current date + time |
| `timedelta(days=, hours=...)` | A duration |
| `d2 - d1` | A `timedelta` between two dates |
| `.strftime(fmt)` | datetime -> string |
| `datetime.strptime(s, fmt)` | string -> datetime |
| `.isoformat()` / `fromisoformat()` | ISO 8601 conversion |

**Format codes:** `%Y` year · `%m` month · `%d` day · `%H` hour · `%M` minute · `%S` second · `%A` weekday · `%B` month name · `%p` AM/PM

```python
from datetime import date, time, datetime, timedelta

today = date.today()
print('today:', today)
print('weekday (Mon=0):', today.weekday())

now = datetime.now()
print('now:', now)
print('date part:', now.date(), '| time part:', now.time())
```

```python
# timedelta: durations and date math
one_week = timedelta(days=7)
print('today + 1 week:', today + one_week)
print('today - 10 days:', today - timedelta(days=10))

delta = date(2026, 12, 31) - today
print('days until 2026-12-31:', delta.days)

span = timedelta(days=2, hours=3, minutes=30)
print('total seconds:', span.total_seconds())
```

```python
# strftime: datetime -> string
sample = datetime(2026, 6, 23, 23, 40, 5)
print(sample.strftime('%Y-%m-%d'))
print(sample.strftime('%d/%m/%Y %H:%M'))
print(sample.strftime('%A, %B %d, %Y'))
print(sample.strftime('%I:%M %p'))
```

```python
# strptime: string -> datetime
parsed = datetime.strptime('2026-06-23 23:40', '%Y-%m-%d %H:%M')
print('parsed:', parsed, '| type:', type(parsed).__name__)

iso = datetime.fromisoformat('2026-06-23T23:40:00')
print('fromisoformat:', iso)
```

```python
# Real-world: age in days
birthday = date(2000, 1, 1)
age_days = (date.today() - birthday).days
print(f'{age_days:,} days old (~{age_days // 365} years)')
```

## Summary

| Concept | Key Point |
|---|---|
| `date` / `time` / `datetime` | dates, clock times, both |
| `datetime.now()` / `date.today()` | current moment |
| `timedelta` | duration; date math returns one |
| `strftime` / `strptime` | format vs parse |
| `fromisoformat` | fast ISO 8601 parse |
