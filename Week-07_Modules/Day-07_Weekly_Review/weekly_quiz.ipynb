## Day 1: Import System & venv


| Concept | Meaning |

| `pip` | `pip install pkg`, `pip freeze > requirements.txt` |

## Day 4: datetime Module

```python
from datetime import date, datetime, timedelta

datetime.now()                       # current date + time
date.today()                         # current date
datetime.now().strftime('%Y-%m-%d')  # format -> string
datetime.strptime('2026-06-24', '%Y-%m-%d')  # parse <- string

now = datetime.now()
now + timedelta(days=7, hours=3)     # arithmetic
(d2 - d1)                            # difference -> timedelta
```

| Code | Meaning | Code | Meaning |
|------|---------|------|---------|
| `%Y` | 4-digit year | `%H` | hour (24h) |
| `%m` | month | `%M` | minute |
| `%d` | day | `%S` | second |

````python
from datetime import datetime, timedelta

now = datetime.now()
print('Now      :', now.strftime('%Y-%m-%d %H:%M:%S'))
print('In 10 days:', (now + timedelta(days=10)).strftime('%Y-%m-%d'))

span = datetime(2026, 6, 24) - datetime(2026, 6, 20)
print('Span     :', span, '->', span.days, 'days')
````

## Day 5: random Module

```python
import random

random.seed(42)            # reproducible sequence
random.random()           # float in [0.0, 1.0)
random.randint(1, 6)      # int in [1, 6] inclusive
random.choice(['a','b'])  # one element
random.choices(pop, k=3)  # k elements, with replacement (weights ok)

```

````python
import random

random.seed(42)
print('randint  :', random.randint(1, 6))
print('choice   :', random.choice(['rock', 'paper', 'scissors']))
print('sample   :', random.sample(range(1, 50), 6))
deck = list('AKQJ')
random.shuffle(deck)
print('shuffled :', deck)
````

## Day 6: Regular Expressions (re)

```python
import re

re.search(r'\d+', text)    # first match anywhere -> Match | None
re.match(r'\d+', text)     # match at START only
re.findall(r'\d+', text)   # all matches -> list
re.sub(r'\d+', '#', text)  # replace
re.split(r'\W+', text)     # split on pattern

m = re.search(r'(?P<user>\w+)@(?P<host>[\w.]+)', 'a@b.com')
m.group('user')            # 'a'
```

| Token | Matches | Token | Matches |
|-------|---------|-------|---------|
| `\d` | digit | `*` | 0 or more |
| `\w` | word char | `+` | 1 or more |
| `\s` | whitespace | `?` | 0 or 1 |
| `.` | any char | `{n,m}` | n to m |
| `^` `$` | start / end | `(...)` | capture group |

**Always use raw strings:** `r"\d+"`

````python
import re

text = 'Contact a@x.com or b@y.org by 2026-06-24.'
print('emails:', re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', text))
print('date  :', re.search(r'\d{4}-\d{2}-\d{2}', text).group())
print('masked:', re.sub(r'\d', '#', text))
````
