# Week 07 - Day 06 | Regular Expressions (`re`)

**Topics:** `search` · `match` · `findall` · `sub` · `split` · groups · `compile` · flags

> Always write patterns as raw strings: `r"\d+"`

---

## Cheatsheet

| Function | Purpose |
|---|---|
| `re.search(p, s)` | First match anywhere |
| `re.match(p, s)` | Match only at the start |
| `re.findall(p, s)` | All matches as a list |
| `re.finditer(p, s)` | All matches as Match objects |
| `re.sub(p, repl, s)` | Replace matches |
| `re.split(p, s)` | Split on a pattern |
| `re.compile(p)` | Reusable compiled pattern |

**Classes:** `\d` digit · `\w` word · `\s` space · `.` any · `[abc]` set · `^` start · `$` end · `` boundary  
**Quantifiers:** `*` 0+ · `+` 1+ · `?` 0/1 · `{n,m}` range · add `?` for non-greedy

```python
import re
text = 'The year 2026 starts now, 100 days to go.'

print('search:', re.search(r'\d+', text).group())
print('match The:', re.match(r'The', text) is not None)
print('findall:', re.findall(r'\d+', text))
print('sub:', re.sub(r'\d+', '#', text))
print('split:', re.split(r'\W+', 'a-b_c d'))
```

```python
# Character classes & quantifiers
print('vowels:', re.findall(r'[aeiou]', 'education'))
print('greedy <.+>:', re.findall(r'<.+>', '<a><b>'))
print('lazy   <.+?>:', re.findall(r'<.+?>', '<a><b>'))
print('3-digit:', re.findall(r'\d{3}', '12 345 6789'))
```

```python
# Groups: numbered and named
m = re.search(r'(\d{4})-(\d{2})-(\d{2})', '2026-06-24')
print('groups:', m.groups(), '| year:', m.group(1))

m = re.search(r'(?P<user>\w+)@(?P<domain>[\w.]+)', 'berke@example.com')
print('named:', m.group('user'), '/', m.group('domain'))

print('backref sub:', re.sub(r'(\w+)-(\w+)', r'-', 'first-second'))
```

```python
# compile + flags
email_re = re.compile(r'[\w.+-]+@[\w-]+\.[\w.-]+')
print(email_re.findall('Mail a@x.com or b@y.org please'))
print('IGNORECASE:', re.findall(r'python', 'Python PYTHON pyThon', re.I))
```

```python
# Practical validator
def is_valid_email(addr):
    return re.match(r'^[\w.+-]+@[\w-]+\.[\w.-]+$', addr) is not None

for a in ['berke@example.com', 'bad@', 'no-at.com', 'a.b@c.co.uk']:
    print(f'{a:<20} -> {is_valid_email(a)}')
```

## Summary

| Concept | Key Point |
|---|---|
| `search` vs `match` | anywhere vs at start |
| `findall` / `finditer` | list vs Match objects |
| `sub` / `split` | replace / split |
| `(...)` / `(?P<n>...)` | capture / named groups; `` backref |
| `compile` | reuse a pattern |
| raw strings | always `r"..."` |
