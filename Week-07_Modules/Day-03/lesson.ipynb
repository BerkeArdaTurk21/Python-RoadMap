## Cheatsheet

| Tool | Purpose |
|---|---|
| `sys.argv` | Raw CLI args (list of strings, `argv[0]` = script) |
| `sys.exit(code)` | Exit; `0` = success, non-zero = error |
| `sys.path` | Module import search path |
| `sys.stdin/stdout/stderr` | Standard streams |
| `argparse.ArgumentParser()` | Build a CLI parser |
| `add_argument()` | Define a positional/option/flag |
| `action='store_true'` | Boolean flag |
| `choices=[...]` | Restrict allowed values |
| `add_subparsers()` | git-style subcommands |

```python
import sys

# sys.argv: all strings, argv[0] is the script name
print('argv:', sys.argv)
print('python:', sys.version.split()[0])
print('platform:', sys.platform)
print('sys.path[0]:', repr(sys.path[0]))
```

```python
# Streams: print() -> stdout by default, redirect with file=
print('normal output')
print('error output', file=sys.stderr)

# sys.exit(0) = success, non-zero = error (commented so the cell runs)
# sys.exit('fatal: bad input')  # prints to stderr, exits with code 1
```

```python
import argparse

parser = argparse.ArgumentParser(prog='greet', description='Greet a user.')
parser.add_argument('name', help='who to greet')
parser.add_argument('-n', '--times', type=int, default=1, help='repeat count')
parser.add_argument('--upper', action='store_true', help='uppercase')
parser.add_argument('--lang', choices=['en', 'tr', 'es'], default='en')

# Pass an explicit list so the notebook runs without real CLI input
args = parser.parse_args(['Berke', '-n', '2', '--upper', '--lang', 'tr'])
print(args)
```

```python
greetings = {'en': 'Hello', 'tr': 'Merhaba', 'es': 'Hola'}
msg = f"{greetings[args.lang]}, {args.name}!"
if args.upper:
    msg = msg.upper()
for _ in range(args.times):
    print(msg)
```

```python
# Subcommands (git-style)
tool = argparse.ArgumentParser(prog='taskman')
sub = tool.add_subparsers(dest='command', required=True)

add_p = sub.add_parser('add')
add_p.add_argument('title')
add_p.add_argument('--priority', choices=['low', 'high'], default='low')

list_p = sub.add_parser('list')
list_p.add_argument('--all', action='store_true')

parsed = tool.parse_args(['add', 'Write lesson', '--priority', 'high'])
print('command:', parsed.command)
if parsed.command == 'add':
    print(f"  add '{parsed.title}' (priority={parsed.priority})")
```

## Summary

| Concept | Key Point |
|---|---|
| `sys.argv` | Raw args, all strings, `argv[0]` = script |
| `sys.exit(code)` | `0` success, non-zero error |
| `argparse` | Real CLIs: positionals, options, flags, choices |
| `parse_args()` | Returns a `Namespace` of values |
| `add_subparsers()` | git-style subcommands |
| `--help` / `--version` | Generated automatically |
