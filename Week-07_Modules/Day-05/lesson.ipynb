# Week 07 - Day 05 | The random Module

**Topics:** `random` · `uniform` · `randint` · `choice` · `choices` · `sample` · `shuffle` · `seed`

---

## Cheatsheet

| Function | Purpose |
|---|---|
| `random()` | Float in `[0.0, 1.0)` |
| `uniform(a, b)` | Float in `[a, b]` |
| `randint(a, b)` | Int in `[a, b]` (both inclusive) |
| `randrange(start, stop, step)` | Int like `range()` (end exclusive) |
| `choice(seq)` | One random element |
| `choices(seq, k=)` | k picks WITH replacement (weights ok) |
| `sample(seq, k)` | k UNIQUE picks (no replacement) |
| `shuffle(list)` | Reorder a list in place |
| `seed(n)` | Make the sequence reproducible |

> **Not secure** — use the `secrets` module for passwords/tokens.

```python
import random

print('random():', round(random.random(), 4))
print('uniform(1,10):', round(random.uniform(1, 10), 4))
print('randint(1,6):', random.randint(1, 6))
print('randrange(0,100,5):', random.randrange(0, 100, 5))
```

```python
colors = ['red', 'green', 'blue', 'yellow']
print('choice:', random.choice(colors))
print('choices k=3:', random.choices(colors, k=3))
print('weighted:', random.choices(colors, weights=[10,1,1,1], k=5))
print('sample (unique):', random.sample(range(1, 50), 6))
```

```python
# shuffle is IN PLACE (returns None)
cards = ['A', 'K', 'Q', 'J', '10']
random.shuffle(cards)
print('shuffled in place:', cards)

# shuffle WITHOUT mutating: sample the whole thing
original = [1, 2, 3, 4, 5]
copy = random.sample(original, len(original))
print('original kept:', original, '-> copy:', copy)
```

```python
# seed: reproducible randomness
random.seed(42)
first = [random.randint(1, 100) for _ in range(3)]
random.seed(42)
second = [random.randint(1, 100) for _ in range(3)]
print('run 1:', first)
print('run 2:', second)
print('identical?', first == second)
random.seed()  # back to unpredictable
```

```python
# Simulation: roll two dice 10,000 times
random.seed(1)
rolls = [random.randint(1,6) + random.randint(1,6) for _ in range(10_000)]
sevens = rolls.count(7)
print(f'rolled 7: {sevens} times ({sevens/100:.1f}%) — theory ~16.7%')
```

## Summary

| Concept | Key Point |
|---|---|
| `randint(a, b)` | both ends inclusive |
| `choices` vs `sample` | with vs without replacement |
| `shuffle` | in place, returns `None` |
| `seed(n)` | reproducible sequence |
| security | use `secrets`, not `random` |
