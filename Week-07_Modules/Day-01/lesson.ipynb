## Import Cheatsheet

| Syntax | Usage |
|---|---|
| `import math` | Use as `math.sqrt()` |
| `from math import sqrt` | Use as `sqrt()` |
| `from math import sqrt, pi` | Multiple names |
| `import math as m` | Alias module: `m.sqrt()` |
| `from math import sqrt as s` | Alias name: `s()` |
| `from math import *` | Wildcard (avoid) |

````python
import math
from math import sqrt, pi
import datetime as dt

print(math.hypot(3, 4))   # 5.0
print(sqrt(25))            # 5.0
print(dt.date.today())
````

## `__name__` and `__main__`

Every module has a `__name__`:
- Run directly → `__name__ == '__main__'`
- Imported → `__name__ == 'module_filename'`

```python
def main():
    print('Hello!')

if __name__ == '__main__':
    main()   # only runs when this file is the entry point
```

This lets the same file work as both a **script** and an **importable module**.

````python
import sys

print(f'__name__: {__name__}')          # '__main__' in notebook
print(f'Python:   {sys.version.split()[0]}')
print(f'Executable: {sys.executable}')
print(f'\nsys.path[:3]:')
for p in sys.path[:3]:
    print(f'  {p!r}')
````

## Virtual Environments (venv)

```bash
# 1. Create
python -m venv .venv

# 2. Activate
# Windows: .venv\Scripts\Activate.ps1
# macOS/Linux: source .venv/bin/activate

# 3. Install packages
pip install requests

# 4. Snapshot dependencies
pip freeze > requirements.txt

# 5. Recreate environment elsewhere
pip install -r requirements.txt

# 6. Deactivate
deactivate
```

> **Rule:** one venv per project. Never install into the global Python.

## Summary

| Concept | Key Point |
|---|---|
| `import module` | Full module, use with dot |
| `from module import name` | Name directly in scope |
| `import module as alias` | Shorter reference |
| `__name__ == '__main__'` | Guard for entry-point code |
| `python -m venv .venv` | Isolated environment |
| `pip freeze > req.txt` | Snapshot dependencies |
