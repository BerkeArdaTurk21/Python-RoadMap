## 1. What is Scope?

A scope is the region of code where a name is visible. Python uses lexical scoping — visibility is decided by WHERE the name is written, not by who calls the function.

Four scopes, innermost first:
- **L** — Local: inside the current function
- **E** — Enclosing: inside any outer function
- **G** — Global: module level
- **B** — Built-in: names like `print`, `len`, `range`

This order is the **LEGB rule**.

## 2. Local Scope

```python
def greet():
    message = "Hello!"
    print(message)

greet()
# print(message)  # NameError
```

## 3. Global Scope

Functions can READ globals freely.

```python
counter = 0

def show_counter():
    print(f"counter = {counter}")

show_counter()
```

## 4. The `global` Keyword

Writing to a name inside a function creates a LOCAL by default. Use `global` to modify the module-level variable instead.

```python
x = 10

def shadow():
    x = 99
    print(f"  inside shadow: x = {x}")

def modify_global():
    global x
    x = 99
    print(f"  inside modify_global: x = {x}")

shadow()
print(f"after shadow:        x = {x}")
modify_global()
print(f"after modify_global: x = {x}")
```

## 5. Enclosing Scope — Nested Functions

```python
def outer():
    msg = "hi from outer"
    def inner():
        print(msg)
    inner()

outer()
```

## 6. `nonlocal` — Rebinding an Enclosing Variable

```python
def make_counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

c = make_counter()
print(c(), c(), c())
```

## 7. The LEGB Rule in Action

```python
name = "global Berke"

def outer():
    name = "enclosing Berke"
    def inner():
        name = "local Berke"
        print(name)
    inner()
    print(name)

outer()
print(name)
```

## 8. What is a Closure?

A closure is a function that REMEMBERS variables from its enclosing scope, even after the scope has finished executing.

Created automatically when:
1. You define a nested function
2. It references names from the enclosing function
3. The enclosing function RETURNS the nested function

```python
def make_multiplier(factor):
    def multiply(x):
        return x * factor
    return multiply

double = make_multiplier(2)
triple = make_multiplier(3)

print(double(5))
print(triple(5))
print(double.__closure__[0].cell_contents)
print(triple.__closure__[0].cell_contents)
```

## 9. Practical Closures

```python
def make_power(exp):
    def power(base):
        return base ** exp
    return power

square = make_power(2)
cube   = make_power(3)
print(square(7))
print(cube(3))
```

```python
def counter(start=0, step=1):
    value = start
    def tick():
        nonlocal value
        current = value
        value += step
        return current
    return tick

c1 = counter()
print(c1(), c1(), c1())

c2 = counter(start=100, step=10)
print(c2(), c2(), c2())
```

```python
def build_validator(min_len):
    def is_valid(text):
        return isinstance(text, str) and len(text) >= min_len
    return is_valid

valid_username = build_validator(3)
valid_password = build_validator(8)

print(valid_username("Berke"))
print(valid_username("ab"))
print(valid_password("12345"))
print(valid_password("verysecret"))
```

## 10. Common Pitfall — Late Binding in Loops

Closures capture VARIABLES, not their current values.

```python
# BUG
funcs_bad = []
for i in range(5):
    funcs_bad.append(lambda: i)
print([f() for f in funcs_bad])

# FIX — bind current i via default argument
funcs_good = []
for i in range(5):
    funcs_good.append(lambda i=i: i)
print([f() for f in funcs_good])
```

## Recap

- LEGB: Local → Enclosing → Global → Built-in
- Assignment inside a function creates a LOCAL by default
- `global x`   — assign to module-level `x`
- `nonlocal x` — assign to enclosing function's `x`
- Closures = nested functions that capture names from the outer scope
- Watch out for late binding inside loops — use `lambda x=x: ...`
