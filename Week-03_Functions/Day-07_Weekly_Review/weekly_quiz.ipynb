### Q3. *args Type

What does `*args` collect inside a function?
- A) A dictionary of keyword arguments
- B) A tuple of extra positional arguments
- C) A list of all arguments
- D) The first positional argument only

````python
# Answer: B — *args is a TUPLE
def show(*args):
    print(type(args), args)

show(1, 2, 3)  # <class 'tuple'> (1, 2, 3)
````

### Q4. **kwargs

```python
def total(**kwargs):
    return sum(kwargs.values())

print(total(a=1, b=2, c=3))
```

- A) `{'a':1, 'b':2, 'c':3}`
- B) `6`
- C) `(1, 2, 3)`
- D) `TypeError`

````python
# Answer: B
def total(**kwargs):
    return sum(kwargs.values())

print(total(a=1, b=2, c=3))  # 6
````

### Q5. Lambda with Ternary

```python
f = lambda x: x ** 2 if x % 2 == 0 else x
print(f(4))
```

- A) `4`  B) `16`  C) `2`  D) `True`

````python
# Answer: B — 4 is even → 4**2 = 16
f = lambda x: x ** 2 if x % 2 == 0 else x
print(f(4))   # 16
print(f(3))   # 3 (odd → unchanged)
````

### Q6. filter()

```python

print(list(result))
```

- A) `[1, 2, 3]`  B) `[4, 5]`  C) `[True, True]`  D) filter object

````python
# Answer: B
nums = [1, 2, 3, 4, 5]
print(list(filter(lambda x: x > 3, nums)))  # [4, 5]
````

### Q7. LEGB Rule

What is the LEGB rule?
- A) A Python sorting algorithm
- B) **Local → Enclosing → Global → Built-in** — scope lookup order
- C) Lambda → Expression → Global → Builtin
- D) A memory management technique

**Answer: B** — When Python encounters a name, it searches scopes in this order:
1. **L**ocal (inside the current function)
2. **E**nclosing (outer function scopes)
3. **G**lobal (module level)
4. **B**uilt-in (Python builtins like `len`, `print`)

### Q8. Closure

What is a closure?
- A) A function that takes no arguments
- B) **An inner function that remembers variables from its enclosing scope**
- C) A function that returns None
- D) A decorator that closes over a module

````python
# Answer: B — closure example
def make_adder(n):
    def add(x):
        return x + n   # 'n' is remembered from enclosing scope
    return add

add5 = make_adder(5)
print(add5(10))  # 15 — 'n=5' is still remembered
````

### Q9. Recursion Requirement

What is REQUIRED in every recursive function?
- A) A return type annotation
- B) A global variable counter
- C) **A base case**
- D) A loop

**Answer: C** — The base case stops recursion. Without it → `RecursionError: maximum recursion depth exceeded`.

### Q10. Trace the Recursion

```python
def f(n):
    if n == 0:
        return 0
    return n + f(n - 1)

print(f(4))
```

- A) `4`  B) `24`  C) `10`  D) `RecursionError`

````python
# Answer: C — 4+3+2+1+0 = 10
def f(n):
    if n == 0:
        return 0
    return n + f(n - 1)

print(f(4))  # 10
````

### C1. *args + keyword-only default

```python
def describe(*args, sep=', '):
    return sep.join(str(a) for a in args)

print(describe(1, 2, 3))
print(describe('a', 'b', 'c', sep=' - '))
```

What is printed?

````python
# Answer: A) 1, 2, 3  /  a - b - c
def describe(*args, sep=', '):
    return sep.join(str(a) for a in args)

print(describe(1, 2, 3))                    # 1, 2, 3
print(describe('a', 'b', 'c', sep=' - '))   # a - b - c
````

### C2. LEGB Scope Trace

```python
x = 'global'

def outer():
    x = 'enclosing'
    def inner():
        print(x)
    inner()

outer()
print(x)
```

````python
# Answer: B) enclosing / global
x = 'global'

def outer():
    x = 'enclosing'
    def inner():
        print(x)   # finds 'enclosing' in E scope
    inner()

outer()    # enclosing
print(x)   # global
````

### C5. Trace Recursive Output

```python
def mystery(n):
    if n <= 0:
        return ''
    return mystery(n - 1) + str(n)

print(mystery(4))
```

````python
# Answer: B) '1234'
# The recursive call resolves FIRST, so digits build left-to-right
def mystery(n):
    if n <= 0:
        return ''
    return mystery(n - 1) + str(n)

print(mystery(4))  # 1234
````
