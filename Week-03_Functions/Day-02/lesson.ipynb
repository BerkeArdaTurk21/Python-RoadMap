# Week 3 - Day 2: Default & Keyword Arguments
**Topic:** Default values, keyword arguments, argument order

## 1. Default Parameter Values
Assign a default with `=`. Used automatically when the caller omits that argument.

```python
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")

greet("Berke")           # Hello, Berke!   ← default used
greet("Alice", "Hi")     # Hi, Alice!      ← default overridden

def create_user(username, role="member", active=True):
    print(f"User: {username} | Role: {role} | Active: {active}")

create_user("berke")
create_user("admin_user", "admin")
create_user("old_user", "member", False)
```

## 2. Keyword Arguments
Name arguments explicitly when calling — order no longer matters.

```python
def power(base, exponent):
    return base ** exponent

print(power(2, 8))                   # 256 — positional
print(power(exponent=8, base=2))     # 256 — keyword, any order

def connect(host, port, timeout=30, use_ssl=True):
    print(f"{host}:{port}  timeout={timeout}  ssl={use_ssl}")

connect("db.example.com", 5432)
connect("db.example.com", 5432, timeout=60)
connect("db.example.com", 5432, use_ssl=False)
```

## 3. Argument Order Rule
**Required parameters** (no default) must come **before** optional ones (with default).

```python
# ✅ Correct order
def describe(name, age, city="Unknown", country="Unknown"):
    print(f"{name}, {age}, from {city}, {country}")

describe("Berke", 22)
describe("Alice", 30, "London")
describe("Bob",   25, country="USA")

# ❌ SyntaxError — default before required:
# def bad(greeting="Hi", name):  ← not allowed
```

## 4. Keyword-Only Arguments (after `*`)
Parameters after a bare `*` can **only** be passed by keyword.

```python
def send_email(to, subject, *, cc="", bcc="", priority="normal"):
    print(f"To: {to} | Subject: {subject} | Priority: {priority}")
    if cc:  print(f"  CC: {cc}")

send_email("alice@x.com", "Hello")
send_email("bob@x.com",   "Meeting", cc="mgr@x.com")
send_email("team@x.com",  "Alert",   priority="high")
```

## 5. Mutable Default Argument — Common Pitfall
Never use a list or dict as a default. Use `None` instead.

```python
# ❌ BUG — shared list
def append_bad(item, items=[]):
    items.append(item)
    return items

print(append_bad("a"))   # ['a']
print(append_bad("b"))   # ['a', 'b']  ← unexpected!

# ✅ CORRECT — fresh list every call
def append_good(item, items=None):
    if items is None:
        items = []
    items.append(item)
    return items

print(append_good("a"))  # ['a']
print(append_good("b"))  # ['b']
```

## 6. Practical Example — Report Generator

```python
def print_report(title, data, *,
                 separator="─", width=40,
                 show_count=True, show_total=False):
    print(separator * width)
    print(f"  {title.upper():^{width - 4}}")
    print(separator * width)
    for i, value in enumerate(data, 1):
        print(f"  {i:2}. {value}")
    print(separator * width)
    if show_count: print(f"  Count : {len(data)}")
    if show_total: print(f"  Total : {sum(data)}")

scores = [88, 92, 75, 95, 60, 83]
print_report("Student Scores", scores)
print()
print_report("Monthly Sales", [1200, 980, 1500],
             separator="=", show_total=True)
```
