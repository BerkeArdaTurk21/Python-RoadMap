# Week 3 - Day 1: Defining Functions
**Topic:** def, parameters, return, docstrings

## 1. Why Functions?
Functions let you **reuse code** without repeating it (DRY — Don't Repeat Yourself).

````python
def square(n):
    return n ** 2

print(square(2))    # 4
print(square(3))    # 9
print(square(5))    # 25
````

## 2. Defining a Function — def
```python
def function_name(parameters):
    body
    return value   # optional
```

````python
def greet():
    print("Hello, World!")

greet()     # Hello, World!
greet()     # called again
````

## 3. Parameters and Arguments
- **Parameter** = variable name in the definition
- **Argument** = actual value passed when calling

````python
def greet_person(name):         # 'name' is a parameter
    print(f"Hello, {name}!")

greet_person("Berke")           # "Berke" is the argument
greet_person("Alice")

def add(a, b):
    print(f"{a} + {b} = {a + b}")

add(3, 5)
add(10, 20)
````

## 4. return Statement
`return` sends a value back to the caller and exits the function immediately.

````python
def multiply(a, b):
    return a * b

result = multiply(4, 7)
print(result)               # 28

# Return multiple values as a tuple
def min_max(numbers):
    return min(numbers), max(numbers)

low, high = min_max([4, 2, 9, 1, 7])
print(f"Min: {low}, Max: {high}")
````

## 5. Functions Without return
A function with no `return` implicitly returns `None`.

````python
def print_separator(char="-", length=30):
    print(char * length)

print_separator()
print_separator("=", 20)

result = print_separator()
print(result)               # None
````

## 6. Docstrings
A string literal on the **first line of the function body** that documents what the function does.

````python
def celsius_to_fahrenheit(celsius):
    """
    Convert a temperature from Celsius to Fahrenheit.

    Parameters:
        celsius (float): Temperature in degrees Celsius.
    Returns:
        float: Temperature in degrees Fahrenheit.
    """
    return celsius * 9 / 5 + 32

print(celsius_to_fahrenheit(100))   # 212.0
print(celsius_to_fahrenheit.__doc__)
````

## 7. Functions as Values
Functions are first-class objects — assign them to variables, pass as arguments.

````python
def double(x): return x * 2
def triple(x): return x * 3

def apply(func, value):
    return func(value)

print(apply(double, 7))     # 14
print(apply(triple, 7))     # 21
````

## 8. Type Hints
Optional annotations that improve readability without enforcing types at runtime.

````python
def add_ints(a: int, b: int) -> int:
    """Return the sum of two integers."""
    return a + b

def full_name(first: str, last: str) -> str:
    """Return the full name."""
    return f"{first} {last}"

print(add_ints(3, 7))               # 10
print(full_name("Berke", "Türk"))    # Berke Türk
````

## 9. Practical Example — Unit Converter

````python
def km_to_miles(km: float) -> float:
    """Convert kilometres to miles."""
    return km * 0.621371

def kg_to_pounds(kg: float) -> float:
    """Convert kilograms to pounds."""
    return kg * 2.20462

def liters_to_gallons(liters: float) -> float:
    """Convert litres to US gallons."""
    return liters * 0.264172

def convert_and_report(km, kg, liters):
    print("── Unit Conversion Report ──")
    print(f"  {km} km    = {km_to_miles(km):.2f} miles")
    print(f"  {kg} kg    = {kg_to_pounds(kg):.2f} pounds")
    print(f"  {liters} L = {liters_to_gallons(liters):.2f} gallons")

convert_and_report(100, 70, 50)
````
