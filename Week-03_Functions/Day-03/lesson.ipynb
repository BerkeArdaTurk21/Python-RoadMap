# Week 3 - Day 3: *args & **kwargs

**Topic:** Variable length arguments and unpacking

By the end of this lesson you will:
- Accept any number of positional arguments with `*args`
- Accept any number of keyword arguments with `**kwargs`
- Combine them with normal and keyword-only parameters
- Unpack iterables and dicts at the call site with `*` and `**`
- Forward arguments transparently from one function to another

## 1. *args — Variable Number of Positional Args

A parameter prefixed with `*` collects ALL extra positional arguments into a **tuple**.
The name `args` is convention only — `*anything` works.

````python
def total(*numbers):
    print(f"  numbers = {numbers}  (type: {type(numbers).__name__})")
    return sum(numbers)

print(total(1, 2, 3))
print(total(10, 20, 30, 40, 50))
print(total())
````

````python
def announce(prefix, *messages):
    for msg in messages:
        print(f"{prefix}: {msg}")

announce("INFO", "Server started", "Listening on 8080", "Ready")
````

## 2. **kwargs — Variable Number of Keyword Args

A parameter prefixed with `**` collects ALL extra keyword arguments into a **dict**.

````python
def describe(**info):
    print(f"  info = {info}  (type: {type(info).__name__})")
    for key, value in info.items():
        print(f"  {key:<10} -> {value}")

describe(name="Berke", age=22, city="Warsaw")
````

## 3. Combining *args and **kwargs

Order in the signature MUST be:
```
def f(positional, *args, keyword_only=default, **kwargs)
```

````python
def log(level, *messages, timestamp=True, **meta):
    if timestamp:
        from datetime import datetime
        stamp = datetime.now().strftime("%H:%M:%S")
        print(f"[{stamp}] [{level}]", *messages)
    else:
        print(f"[{level}]", *messages)
    if meta:
        for k, v in meta.items():
            print(f"   {k}={v}")

log("INFO", "User logged in", user="berke", ip="192.168.1.1")
log("ERROR", "DB connection failed", timestamp=False, retries=3, host="db.local")
````

## 4. Unpacking with * and ** at the Call Site

- `*`  unpacks an iterable into positional arguments
- `**` unpacks a dict into keyword arguments

````python
def greet(greeting, name, punctuation):
    print(f"{greeting}, {name}{punctuation}")

args_list = ["Hello", "Berke", "!"]
greet(*args_list)

kwargs_dict = {"greeting": "Hi", "name": "Alice", "punctuation": "."}
greet(**kwargs_dict)
````

## 5. Forwarding Arguments — A Key Pattern

`*args` and `**kwargs` let one function transparently pass arguments to another — common in wrappers, decorators, and subclass `__init__` methods.

````python
def debug_call(func, *args, **kwargs):
    print(f"-> Calling {func.__name__}({args}, {kwargs})")
    result = func(*args, **kwargs)
    print(f"<- Returned {result}")
    return result

def add(a, b):
    return a + b

def power(base, exp=2):
    return base ** exp

debug_call(add, 3, 4)
debug_call(power, 2, exp=10)
````

## 6. Flexible Configuration with **kwargs

````python
def build_user(username, **fields):
    user = {"username": username}
    user.update(fields)
    return user

print(build_user("berke"))
print(build_user("alice", age=30, city="London"))
print(build_user("bob", age=25, role="admin", active=True, score=95))
````

## 7. Practical Example — Mini Event Emitter

````python
def emit(event, *receivers, sender="system", **payload):
    print(f"-- Event: {event}  (from {sender}) --")
    for r in receivers:
        print(f"  -> delivered to {r}")
    if payload:
        print("  payload:")
        for k, v in payload.items():
            print(f"     {k} = {v}")

emit("login", "auth_service", "logger",
     sender="berke", ip="10.0.0.5", success=True)

emit("order.created", "billing", "inventory", "email",
     sender="checkout",
     order_id=4821, total=129.99, currency="EUR")
````

## Recap

- `*args`  collects extra positional args into a **tuple**
- `**kwargs` collects extra keyword args into a **dict**
- Signature order: positional → `*args` → keyword-only → `**kwargs`
- `*iterable` and `**dict` at the call site unpack into arguments
- Use them to write wrappers, decorators, and flexible APIs
