## Minimal Custom Exception

```python
class MyError(Exception):
    """Descriptive docstring."""
```

That's it. You can raise, catch, and inspect it just like any built-in exception.

````python
# Hierarchy: catch broad or specific
class AppError(Exception):
    """Base exception."""

class DatabaseError(AppError): pass
class NetworkError(AppError):  pass

for exc_cls, svc in [(DatabaseError, 'db'), (NetworkError, 'net')]:
    try:
        raise exc_cls(f'could not reach {svc}')
    except AppError as e:   # catches both
        print(f'{type(e).__name__}: {e}')
````

````python
# Custom attributes
class ValidationError(AppError):
    def __init__(self, message, field, value):
        super().__init__(message)
        self.field = field
        self.value = value

    def __str__(self):
        return f'[{self.field}={self.value!r}] {super().__str__()}'

try:
    raise ValidationError('must not be empty', field='email', value='')
except ValidationError as e:
    print(e)
    print(f'  field={e.field}, value={e.value!r}')
````

## When Custom vs Built-in?

| Use built-in | Use custom |
|---|---|
| `ValueError` fits semantically | Need extra data (field, code, status) |
| One-off error | Part of a catchable hierarchy |
| Library caller expects standard types | Building a public library API |

**Rule of thumb:** if you'd write `except ValueError` and check `str(e)` to distinguish errors — time for a custom class.

## Summary

```python
class MyError(Exception):          # inherit from Exception
    def __init__(self, msg, code):
        super().__init__(msg)       # pass message to base
        self.code = code            # store extra data
```

- Name ends in `Error`
- Inherit from `Exception`, not `BaseException`
- One hierarchy per subsystem
- `super().__init__(msg)` so `str(e)` works
