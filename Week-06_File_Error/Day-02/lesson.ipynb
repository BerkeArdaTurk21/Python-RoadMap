## File Modes Cheatsheet

| Mode | Creates? | Truncates? | Position |
|------|----------|------------|----------|
| `'w'` | yes | **yes** | start |
| `'a'` | yes | no | **end** |
| `'x'` | only new | no | start |
| `'r+'` | no | no | start |

```python
import os
# write() — creates or overwrites
with open('out.txt', 'w', encoding='utf-8') as f:
    f.write('Line 1\nLine 2\n')

# append — adds to end
with open('out.txt', 'a', encoding='utf-8') as f:
    f.write('Line 3\n')

with open('out.txt') as f:
    print(f.read())
os.remove('out.txt')
```

```python
# writelines — no auto newlines!
with open('fruits.txt', 'w', encoding='utf-8') as f:
    f.writelines(['apple\n', 'banana\n', 'cherry\n'])
with open('fruits.txt') as f: print(f.read())
os.remove('fruits.txt')
```

```python
import json
data = {'name': 'Alice', 'scores': [90, 85, 92]}
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=2)

with open('data.json') as f:
    loaded = json.load(f)
print(loaded['name'], loaded['scores'])
os.remove('data.json')
```

## Summary

| Method | Does |
|--------|------|
| `'w'` mode | Create or OVERWRITE |
| `'a'` mode | Append to end |
| `'x'` mode | Create only (error if exists) |
| `write(str)` | Writes one string |
| `writelines(it)` | Writes each string in iterable (no auto `\n`) |
| `json.dump(obj, f)` | Serialize Python → JSON file |
| `json.load(f)` | Deserialize JSON file → Python |
