## Days 1-2: File I/O Cheatsheet

| Mode | Creates | Truncates | Position |
|------|---------|-----------|----------|
| `'r'` | no | no | start |
| `'w'` | yes | **yes** | start |
| `'a'` | yes | no | end |
| `'x'` | only new | no | start |
| `'r+'` | no | no | start |

```python
# Always use 'with' — auto-closes
with open('file.txt', encoding='utf-8') as f:
    content = f.read()       # whole file as string
    lines   = f.readlines()  # list of lines

import json
with open('data.json', 'w') as f:
    json.dump(obj, f, indent=2)   # write
with open('data.json') as f:
    obj = json.load(f)            # read
```

````python
import os, json
from pathlib import Path

# Write and read back
with open('demo.txt', 'w', encoding='utf-8') as f:
    f.writelines(f'Line {i}\n' for i in range(1, 4))

with open('demo.txt', encoding='utf-8') as f:
    for line in f:
        print(line, end='')

os.remove('demo.txt')
````

## Days 4-5: Exception Handling Patterns

```python
# Full structure
try:
    result = risky()
except ValueError as e:
    handle(e)
except (TypeError, KeyError):
    handle_type_or_key()
else:
    use(result)      # only if no exception
finally:
    cleanup()        # always

# Raise patterns
raise ValueError('bad value')          # signal error
raise                                   # re-raise (in except)
raise RuntimeError('...') from e        # chain
raise RuntimeError('...') from None     # suppress chain
assert cond, 'msg'                      # internal invariant
```

````python
# Exception handling demo
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('  Cannot divide by zero')
    else:
        print(f'  Result: {result}')
    finally:
        print('  Done.')

safe_divide(10, 2)
safe_divide(10, 0)
````

## Day 6: Custom Exception Pattern

```python
class AppError(Exception):
    """Base exception for the application."""

class ValidationError(AppError):
    def __init__(self, message, field, value):
        super().__init__(message)   # message shows in str(e)
        self.field = field
        self.value = value

    def __str__(self):
        return f'[{self.field}={self.value!r}] {super().__str__()}'

# Catch broad:
except AppError as e:
    ...
# Or specific:
except ValidationError as e:
    print(e.field, e.value)
```

### Best Practices
- Name ends in `Error`
- Inherit from `Exception` (not `BaseException`)
- Call `super().__init__(msg)`
- Store extra data as attributes
- One hierarchy per subsystem
