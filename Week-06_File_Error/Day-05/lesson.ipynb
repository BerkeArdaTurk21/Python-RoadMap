# Week 06 - Day 05 | Raising Exceptions

**Topics:** `raise` · `raise from` · bare `raise` · `assert` · choosing exception types

---

## When to raise

| Scenario | What to raise |
|---|---|
| Wrong value for correct type | `ValueError` |
| Wrong type entirely | `TypeError` |
| Must be overridden in subclass | `NotImplementedError` |
| Generic runtime failure | `RuntimeError` |
| Internal invariant broken | `AssertionError` (via `assert`) |
| File doesn't exist | `FileNotFoundError` |

```python
# raise — signal an error
def set_age(age):
    if age < 0 or age > 150:
        raise ValueError(f'Age must be 0-150, got {age}')
    return age

try:
    set_age(200)
except ValueError as e:
    print(f'Caught: {e}')
```

```python
# Bare raise — re-raise inside except
def load(path):
    try:
        open(path)
    except FileNotFoundError:
        print('[log] file missing')
        raise   # re-raises with original traceback

try:
    load('missing.txt')
except FileNotFoundError as e:
    print(f'Caller sees: {e}')
```

```python
# raise B from A — exception chaining
def parse(path):
    try:
        open(path)
    except FileNotFoundError as e:
        raise RuntimeError(f'Config not found: {path}') from e

try:
    parse('config.json')
except RuntimeError as e:
    print(f'RuntimeError: {e}')
    print(f'Caused by: {type(e.__cause__).__name__}')
```

```python
# assert — internal invariants only
def average(nums):
    assert isinstance(nums, list), 'nums must be a list'
    assert len(nums) > 0, 'cannot average empty list'
    return sum(nums) / len(nums)

print(average([10, 20, 30]))   # 20.0
try:
    average([])
except AssertionError as e:
    print(f'AssertionError: {e}')
```

## Summary

| Syntax | Purpose |
|---|---|
| `raise ValueError('msg')` | Signal a bad value |
| `raise` | Re-raise current exception (in except) |
| `raise B from A` | Wrap A in B; both visible in traceback |
| `raise B from None` | Suppress original cause in traceback |
| `assert cond, 'msg'` | Internal invariant (disabled with -O) |
