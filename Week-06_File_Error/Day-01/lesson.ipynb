# Week 06 - Day 01 | Reading Files

**Topics:** `open()` · file modes · `read()` · `readline()` · `readlines()` · `with` statement · `seek()` · `tell()`

---

## 1. File Modes

| Mode | Meaning |
|------|---------|
| `'r'` | Read (default) — file must exist |
| `'w'` | Write — creates or **overwrites** |
| `'a'` | Append — creates or adds to end |
| `'x'` | Exclusive create — fails if exists |
| `'b'` | Binary (combine: `'rb'`, `'wb'`) |

Always specify `encoding='utf-8'`.

```python
# Create a sample file
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write('Line 1: Hello\nLine 2: Python\nLine 3: Files\n')

# Read entire file
with open('sample.txt', 'r', encoding='utf-8') as f:
    content = f.read()
print(content)
print('Closed:', f.closed)  # True — auto-closed by 'with'
```

## 2. read() vs readline() vs readlines()

```python
# readline() — one line at a time
with open('sample.txt', 'r', encoding='utf-8') as f:
    print(repr(f.readline()))   # 'Line 1: Hello\n'
    print(repr(f.readline()))   # 'Line 2: Python\n'

# readlines() — list of all lines
with open('sample.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
print(lines)
```

## 3. Iterating — Most Memory-Efficient

For large files, iterate line by line instead of loading everything at once.

```python
with open('sample.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line.rstrip())   # strip trailing newline
```

## 4. seek() and tell()

```python
with open('sample.txt', 'r', encoding='utf-8') as f:
    print(f.tell())     # 0 — at start
    f.read(6)           # move cursor 6 bytes
    print(f.tell())     # 6
    f.seek(0)           # back to start
    print(f.readline().rstrip())  # Line 1: Hello

import os; os.remove('sample.txt')
```

## Summary

| Method | Does |
|--------|------|
| `f.read()` | Entire file as one string |
| `f.readline()` | One line (keeps `\n`) |
| `f.readlines()` | List of all lines |
| `for line in f:` | Lazy — memory efficient |
| `f.seek(n)` | Jump to byte position n |
| `f.tell()` | Current byte position |
