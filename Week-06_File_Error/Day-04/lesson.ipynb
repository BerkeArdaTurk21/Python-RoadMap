## Structure of a try Block

```python
try:
    # guarded code
except SpecificError:
    # handle it
except (TypeA, TypeB) as e:
    # handle either, 'e' is the exception
else:
    # runs ONLY if no exception occurred
finally:
    # ALWAYS runs — use for cleanup
```

````python
# Basic try/except
try:
    x = int('abc')
except ValueError as e:
    print(f'Caught: {e}')

# Multiple except blocks
data = [10, 20, 30]
for val in ['1', 'abc', '99']:
    try:
        print(data[int(val)])
    except ValueError:
        print(f"'{val}' is not an integer")
    except IndexError:
        print(f"Index {val} out of range")
````

````python
# else and finally
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print('  except: cannot divide by zero')
    else:
        print(f'  else:   {a}/{b} = {result}')
    finally:
        print('  finally: done')

print('divide(10, 2):')
divide(10, 2)
print('divide(10, 0):')
divide(10, 0)
````

## Exception Hierarchy (Partial)

```
BaseException
└── Exception
    ├── ValueError
    ├── TypeError
    ├── IndexError
    ├── KeyError
    ├── AttributeError
    └── OSError
        ├── FileNotFoundError
        └── PermissionError
```

Catching `OSError` also catches `FileNotFoundError` (it's a subclass).
Always put **most specific** exceptions first.

## Summary

| Clause | When it runs |
|---|---|
| `try:` | Always (guarded block) |
| `except E:` | When E (or subclass) is raised |
| `except E as e:` | Same + `e` holds the exception |
| `except (A, B):` | When A or B is raised |
| `else:` | Only if NO exception in try |
| `finally:` | Always — exception or not |
