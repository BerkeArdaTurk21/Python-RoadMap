## os.path (Classic Approach)

| Function | Returns |
|---|---|
| `os.path.join(a, b)` | Platform-correct joined path |
| `os.path.basename(p)` | Filename only |
| `os.path.dirname(p)` | Directory only |
| `os.path.splitext(p)` | `(base, extension)` tuple |
| `os.path.exists(p)` | `True` if path exists |
| `os.path.isfile(p)` | `True` if regular file |
| `os.path.isdir(p)` | `True` if directory |

```python
import os

p = '/Users/alice/projects/app/main.py'
print(os.path.basename(p))    # main.py
print(os.path.dirname(p))     # /Users/alice/projects/app
print(os.path.splitext(p))    # ('/Users/alice/projects/app/main', '.py')

joined = os.path.join('data', '2024', 'report.csv')
print(joined)                 # data/2024/report.csv (or data\2024\report.csv on Windows)
```

## pathlib.Path — Modern OOP Approach

`pathlib` was added in Python 3.4 and is now the preferred way to work with paths.
The `/` operator joins path segments — much cleaner than `os.path.join()`.

```python
from pathlib import Path

full = Path('/Users/alice/projects/app/main.py')
print(full.name)    # main.py
print(full.stem)    # main
print(full.suffix)  # .py
print(full.parent)  # /Users/alice/projects/app

# / operator for joining
data_dir = Path('data') / '2024' / 'report.csv'
print(data_dir)
```

```python
import shutil

# Create dirs and files
tmp = Path('nb_demo')
tmp.mkdir(exist_ok=True)
(tmp / 'sub').mkdir(exist_ok=True)

for name in ['a.py', 'b.py', 'c.txt']:
    (tmp / name).write_text('x', encoding='utf-8')

# List contents
print('All items:')
for item in sorted(tmp.iterdir()):
    print(f'  {item.name}  (dir={item.is_dir()})')

# Glob .py files
print('\nOnly .py files:')
for f in sorted(tmp.glob('*.py')):
    print(f'  {f.name}')

shutil.rmtree(tmp)
```

## Summary

| | `os.path` | `pathlib.Path` |
|---|---|---|
| Join | `os.path.join(a,b,c)` | `Path(a) / b / c` |
| Name only | `os.path.basename(p)` | `p.name` |
| Extension | `os.path.splitext(p)[1]` | `p.suffix` |
| Exists? | `os.path.exists(p)` | `p.exists()` |
| Make dir | `os.makedirs(p)` | `p.mkdir(parents=True)` |
| Find files | `glob.glob(pattern)` | `p.glob("*.py")` |

> **Recommendation:** prefer `pathlib` for all new code — it's object-oriented, readable, and cross-platform.
