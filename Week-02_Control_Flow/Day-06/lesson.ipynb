**Topic:** Python 3.10+ structural pattern matching, guard clauses

> Requires Python 3.10 or higher.

## 1. Basic match / case
Compares a value against multiple patterns and runs the **first matching branch**.

```python
command = "quit"

match command:
    case "start":
        print("Starting...")
    case "stop":
        print("Stopping...")
    case "quit":
        print("Goodbye!")       # This runs
    case _:
        print("Unknown command.")
```

## 2. _ (Wildcard) — The Default Case
`case _` is the catch-all, like `else` in if/elif/else.

```python
status_code = 404

match status_code:
    case 200: print("OK")
    case 404: print("Not Found")    # This runs
    case 500: print("Server Error")
    case _:   print(f"Unknown: {status_code}")
```

## 3. Matching Multiple Values with |
Use `|` to match several values in one `case` branch.

```python
day = "Saturday"

match day:
    case "Monday" | "Tuesday" | "Wednesday" | "Thursday" | "Friday":
        print("Weekday")
    case "Saturday" | "Sunday":
        print("Weekend")    # This runs
    case _:
        print("Invalid day")
```

## 4. Guard Clauses — if Conditions Inside case
Add an `if` guard after a pattern for extra filtering.

```python
def classify_number(n):
    match n:
        case x if x < 0:
            print(f"{x} is negative")
        case 0:
            print("Zero")
        case x if x % 2 == 0:
            print(f"{x} is positive and even")
        case x:
            print(f"{x} is positive and odd")

classify_number(-5)   # negative
classify_number(0)    # Zero
classify_number(8)    # even
classify_number(7)    # odd
```

## 5. Matching Sequences (Tuples / Lists)
match can **destructure** sequences into their parts.

```python
def describe_point(point):
    match point:
        case (0, 0):    print("Origin")
        case (x, 0):    print(f"On x-axis at x={x}")
        case (0, y):    print(f"On y-axis at y={y}")
        case (x, y):    print(f"Point at ({x}, {y})")

describe_point((0, 0))
describe_point((5, 0))
describe_point((0, 3))
describe_point((4, 7))
```

## 6. Matching Lists by Length using *rest

```python
def describe_list(lst):
    match lst:
        case []:            print("Empty list")
        case [x]:           print(f"Single item: {x}")
        case [x, y]:        print(f"Two items: {x} and {y}")
        case [first, *rest]:print(f"First: {first}, rest: {rest}")

describe_list([])
describe_list([42])
describe_list([1, 2])
describe_list([1, 2, 3, 4])
```

## 7. Matching Dictionaries

```python
def handle_event(event):
    match event:
        case {"type": "click",    "button": button}:   print(f"Click: {button}")
        case {"type": "keypress", "key": key}:          print(f"Key: {key}")
        case {"type": "scroll",   "direction": d}:      print(f"Scroll {d}")
        case _:                                          print("Unknown event")

handle_event({"type": "click",    "button": "left"})
handle_event({"type": "keypress", "key": "Enter"})
handle_event({"type": "resize"})
```

## 8. Practical Example — Command Parser

```python
def parse_command(user_input):
    parts = user_input.strip().split()
    match parts:
        case ["quit"] | ["exit"]:
            return "Exiting program."
        case ["help"]:
            return "Commands: quit, help, go <dir>, take <item>"
        case ["go", direction] if direction in ("north", "south", "east", "west"):
            return f"Moving {direction}..."
        case ["go", direction]:
            return f"Cannot go '{direction}'."
        case ["take", item]:
            return f"You picked up: {item}"
        case ["drop", item]:
            return f"You dropped: {item}"
        case []:
            return "Please enter a command."
        case _:
            return f"Unknown: '{user_input}'"

for cmd in ["go north", "take sword", "go up", "help", "quit", "fly"]:
    print(f"> {cmd}  →  {parse_command(cmd)}")
```
