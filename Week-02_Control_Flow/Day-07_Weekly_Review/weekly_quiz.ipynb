## Part 2 — Code Challenges: Answers & Explanations

````python
# Challenge 1 — odd squares
result = []
for i in range(1, 6):
    if i % 2 != 0:
        result.append(i * i)
print(result)   # [1, 9, 25]
````

````python
# Challenge 2 — doubling while
n = 1
while n <= 16:
    n *= 2
print(n)        # 32
````

````python
# Challenge 3 — for/else with break
for i in range(5):
    if i == 3:
        break
else:
    print('done')   # skipped — break was hit
print('end')        # always prints
````

````python
# Challenge 4 — nested loop sum of evens
matrix = [[1,2],[3,4],[5,6]]
total = 0
for row in matrix:
    for val in row:
        if val % 2 == 0:
            total += val
print(total)    # 12  (2+4+6)
````

````python
# Challenge 5 — enumerate search
items = ['a', 'b', 'c', 'd']
for i, item in enumerate(items):
    if item == 'c':
        print(f'found at {i}')   # found at 2
        break
````

## Week 2 — Key Concepts Summary

```python
# if / elif / else
if x > 0:       print('positive')
elif x == 0:    print('zero')
else:           print('negative')

# for loop
for i in range(5):          print(i)
for k, v in d.items():      print(k, v)
for i, v in enumerate(lst): print(i, v)
for a, b in zip(l1, l2):    print(a, b)

# while loop
while condition:
    ...
    if done: break

# break / continue / pass
break      # exit loop entirely
continue   # skip current iteration
pass       # do nothing (placeholder)

# nested loops
for i in range(rows):
    for j in range(cols):
        print(matrix[i][j])

# match / case (Python 3.10+)
match value:
    case 1:             ...
    case 2 | 3:         ...
    case x if x > 10:   ...
    case (x, y):        ...
    case _:             ...   # default
```
