## 1. What Is Inheritance?

Inheritance lets one class (**child/subclass**) reuse and extend the code of another class (**parent/superclass**).

| Term | Meaning |
|------|---------|
| Parent / Superclass | The class being inherited from |
| Child / Subclass | The class that inherits |
| IS-A relationship | A Dog **is an** Animal |

```python
class Child(Parent):
    ...
```

````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def eat(self):
        return f'{self.name} is eating.'

    def __str__(self):
        return f'{self.name} (age {self.age})'


class Dog(Animal):
    def bark(self):
        return f'{self.name} says: Woof!'


rex = Dog('Rex', 3)
print(rex.eat())    # inherited from Animal
print(rex)          # inherited __str__ from Animal
print(rex.bark())   # defined in Dog
````

## 2. `super()` — Calling the Parent

`super()` gives access to the **parent class**.  
Most commonly used in `__init__` to avoid re-typing shared attributes.

````python
class Cat(Animal):
    def __init__(self, name, age, indoor):
        super().__init__(name, age)   # call Animal.__init__
        self.indoor = indoor          # Cat-specific attribute

    def speak(self):
        return f'{self.name} says: Meow!'

    def __str__(self):
        location = 'indoor' if self.indoor else 'outdoor'
        return f'{self.name} ({location} cat, age {self.age})'


whiskers = Cat('Whiskers', 4, indoor=True)
print(whiskers)           # Whiskers (indoor cat, age 4)
print(whiskers.eat())     # inherited
print(whiskers.speak())   # Cat's own method
````

## 3. Method Overriding

A child class can **redefine** a method from the parent.  
Python always calls the **most specific** (child) version first.

````python
class Animal2:
    def speak(self):
        return '...'

class Dog2(Animal2):
    def speak(self):    # overrides Animal2.speak
        return 'Woof!'

class Cat2(Animal2):
    def speak(self):    # overrides Animal2.speak
        return 'Meow!'

class Fish(Animal2):
    pass                # uses Animal2.speak unchanged

for animal in [Dog2(), Cat2(), Fish()]:
    print(animal.speak())   # Woof! / Meow! / ...
````

## 4. Extending a Parent Method with `super()`

Use `super().method()` inside an override to **reuse** the parent logic and add to it.

````python
class Vehicle:
    def __init__(self, make, model, year):
        self.make  = make
        self.model = model
        self.year  = year

    def describe(self):
        return f'{self.year} {self.make} {self.model}'


class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_kwh):
        super().__init__(make, model, year)
        self.battery_kwh = battery_kwh

    def describe(self):
        base = super().describe()         # reuse parent's describe()
        return f'{base} [Electric, {self.battery_kwh} kWh]'


v  = Vehicle('Toyota', 'Camry', 2020)
ev = ElectricCar('Tesla', 'Model 3', 2023, 75)

print(v.describe())    # 2020 Toyota Camry
print(ev.describe())   # 2023 Tesla Model 3 [Electric, 75 kWh]
````

## 5. `isinstance()` and `issubclass()`

- `isinstance(obj, Class)` → `True` if `obj` is an instance of `Class` **or any subclass**
- `issubclass(Child, Parent)` → `True` if `Child` is a subclass of `Parent`

````python
class Shape:     pass
class Rectangle(Shape):  pass
class Square(Rectangle): pass

r = Rectangle()
s = Square()

print(isinstance(r, Rectangle))   # True
print(isinstance(r, Shape))       # True  — Rectangle IS-A Shape
print(isinstance(r, Square))      # False

print(isinstance(s, Square))      # True
print(isinstance(s, Rectangle))   # True  — Square IS-A Rectangle
print(isinstance(s, Shape))       # True  — transitive

print(issubclass(Square, Rectangle))  # True
print(issubclass(Rectangle, Shape))   # True
print(issubclass(Shape, Rectangle))   # False — wrong direction
````

## 6. Multi-level Inheritance

A chain: `A → B → C`  
Python searches methods using the **Method Resolution Order (MRO)**.

````python
class LivingThing:
    def breathe(self):
        return 'Breathing...'

class Animal3(LivingThing):
    def __init__(self, name):
        self.name = name
    def eat(self):
        return f'{self.name} is eating.'

class Dog3(Animal3):
    def bark(self):
        return f'{self.name} says: Woof!'


buddy = Dog3('Buddy')
print(buddy.breathe())   # from LivingThing
print(buddy.eat())       # from Animal3
print(buddy.bark())      # from Dog3

# MRO — the order Python searches for methods
print([c.__name__ for c in Dog3.__mro__])
# ['Dog3', 'Animal3', 'LivingThing', 'object']
````

## 7. Multiple Inheritance (Brief)

Python allows inheriting from **more than one** parent class.  
Use sparingly — it can make code harder to reason about.

````python
class Flyable:
    def fly(self):
        return f'{self.name} is flying!'

class Swimmable:
    def swim(self):
        return f'{self.name} is swimming!'

class Duck(Animal3, Flyable, Swimmable):
    def speak(self):
        return f'{self.name} says: Quack!'


donald = Duck('Donald')
print(donald.eat())    # from Animal3
print(donald.fly())    # from Flyable
print(donald.swim())   # from Swimmable
print(donald.speak())  # from Duck
````

## Summary

| Concept | Key Point |
|---------|-----------|
| Inheritance | Child class reuses parent's code (IS-A relation) |
| `super()` | Calls the parent class method / `__init__` |
| Method override | Child redefines a method from the parent |
| `isinstance()` | True for the class AND all parent classes |
| `issubclass()` | Checks class-level IS-A relationship |
| MRO | Order Python searches for a method |
| Multi-level | `A → B → C` chain of inheritance |
| Multiple | Child inherits from more than one parent |
