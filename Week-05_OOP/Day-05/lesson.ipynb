## 1. Method Overriding

Same method name → different behaviour depending on the object's type.

Python decides **which** method to call at **runtime** based on the actual object.

```python
class Animal:
    def __init__(self, name): self.name = name
    def speak(self): return f'{self.name}: ...'

class Dog(Animal):
    def speak(self): return f'{self.name}: Woof!'

class Cat(Animal):
    def speak(self): return f'{self.name}: Meow!'

for a in [Dog('Rex'), Cat('Whiskers'), Animal('?')]:
    print(a.speak())   # same call — different result
```

## 2. Duck Typing

> *"If it walks like a duck and quacks like a duck, it IS a duck."*

Python checks for **method existence**, not inheritance.

```python
def make_sound(animal):
    print(animal.speak())

class Robot:
    def __init__(self, model): self.name = model
    def speak(self): return f'{self.name}: BEEP BOOP'

make_sound(Dog('Buddy'))    # Buddy: Woof!
make_sound(Robot('R2D2'))   # R2D2: BEEP BOOP — no Animal inheritance needed!
```

## 3. Polymorphic Functions

Functions that work with **any** object having the right interface.

```python
import math

class Shape:
    def area(self): return 0.0
    def describe(self):
        return f'{type(self).__name__} area={self.area():.2f}'

class Rectangle(Shape):
    def __init__(self, w, h): self.w, self.h = w, h
    def area(self): return self.w * self.h

class Circle(Shape):
    def __init__(self, r): self.r = r
    def area(self): return round(math.pi * self.r**2, 2)

shapes = [Rectangle(4, 6), Circle(3)]
for s in shapes:
    print(s.describe())
print(f'Total area: {sum(s.area() for s in shapes):.2f}')
```

## 4. Operator Overloading (Preview)

Operators like `+`, `==`, `<` are polymorphic via **magic methods**.

```python
class Vector:
    def __init__(self, x, y): self.x, self.y = x, y
    def __add__(self, other): return Vector(self.x+other.x, self.y+other.y)
    def __eq__(self, other):  return self.x==other.x and self.y==other.y
    def __str__(self):        return f'Vector({self.x}, {self.y})'

v1 = Vector(1, 2)
v2 = Vector(3, 4)
print(v1 + v2)          # Vector(4, 6)
print(v1 == v2)         # False
print(v1 == Vector(1,2)) # True
```

## Summary

| Concept | Key Point |
|---------|----------|
| Polymorphism | Same interface → different behaviour per type |
| Method overriding | Subclass redefines a parent method |
| Duck typing | Works if the object has the needed methods |
| Polymorphic fn | Works with many types via a shared interface |
| Operator overload | `+`, `==`, `<` call magic methods |
