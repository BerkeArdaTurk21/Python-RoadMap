## 1. `__str__` vs `__repr__`

| Method | Called by | Purpose |
|--------|----------|---------|
| `__str__` | `print()`, `str()` | Human-readable |
| `__repr__` | `repr()`, shell | Developer-readable, should recreate the object |

```python
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y
    def __str__(self):  return f'({self.x}, {self.y})'
    def __repr__(self): return f'Point({self.x!r}, {self.y!r})'

p = Point(3, 4)
print(str(p))    # (3, 4)      — __str__
print(repr(p))   # Point(3, 4) — __repr__
```

## 2. `__len__` and `__bool__`

```python
class Bag:
    def __init__(self):    self._items = []
    def add(self, item):   self._items.append(item)
    def __len__(self):     return len(self._items)
    def __bool__(self):    return len(self._items) > 0

bag = Bag()
print(len(bag), bool(bag))   # 0 False
bag.add('apple')
print(len(bag), bool(bag))   # 1 True
if bag: print('Has items!')
```

## 3. Comparison Methods

```python
class Temperature:
    def __init__(self, c): self.celsius = c
    def __eq__(self, o):  return self.celsius == o.celsius
    def __lt__(self, o):  return self.celsius < o.celsius
    def __repr__(self):   return f'Temperature({self.celsius})'

t1, t2 = Temperature(100), Temperature(0)
print(t1 > t2)   # True
print(sorted([t1, t2, Temperature(50)]))
```

## 4. Arithmetic Methods

```python
class Vector:
    def __init__(self, x, y):  self.x, self.y = x, y
    def __add__(self, o):  return Vector(self.x+o.x, self.y+o.y)
    def __mul__(self, s):  return Vector(self.x*s, self.y*s)
    def __rmul__(self, s): return self.__mul__(s)
    def __repr__(self):    return f'Vector({self.x}, {self.y})'

v1, v2 = Vector(1,2), Vector(3,4)
print(v1 + v2)   # Vector(4, 6)
print(v1 * 3)    # Vector(3, 6)
print(3 * v1)    # Vector(3, 6) — uses __rmul__
```

## 5. `__contains__`, `__getitem__`, `__iter__`

```python
class Playlist:
    def __init__(self, name):
        self.name = name; self._songs = []
    def add(self, song): self._songs.append(song)
    def __contains__(self, song): return song in self._songs
    def __getitem__(self, i):     return self._songs[i]
    def __len__(self):            return len(self._songs)
    def __iter__(self):           return iter(self._songs)

pl = Playlist('Chill')
pl.add('Bohemian Rhapsody'); pl.add('Hotel California')
print('Bohemian Rhapsody' in pl)  # True
print(pl[0])                       # Bohemian Rhapsody
for song in pl: print(song)
```

## Summary

| Method | Triggered by |
|--------|--------------|
| `__str__` | `print(obj)`, `str(obj)` |
| `__repr__` | `repr(obj)`, shell |
| `__len__` | `len(obj)` |
| `__bool__` | `bool(obj)`, `if obj:` |
| `__eq__` | `==` |
| `__lt__` | `<` (enables `sorted()`) |
| `__add__` | `+` |
| `__contains__` | `x in obj` |
| `__getitem__` | `obj[i]` |
| `__iter__` | `for x in obj:` |
